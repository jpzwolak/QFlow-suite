```python
import os, h5py
import matplotlib.pyplot as plt
```

```python
h5_save_file = os.path.join('Data','example_noisy_data.hdf5')

# the best way to use hdf5 files are with context managers to automatically 
# close the file object. 'r' makes it only read the file
with h5py.File(h5_save_file, 'r') as h5f:
    # hdf5 file can be iterated like a dict
    for f, d in h5f.items():
        plt.pcolor(
            d['V_P1_vec'][()], d['V_P2_vec'][()], d['output']['sensor'][()], 
            shading='auto')
        plt.gca().set_aspect(1.)
        plt.title('Noise level: %0.2f'%d['noise_level'][()])
        plt.show()
```
